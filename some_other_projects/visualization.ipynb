```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('dataupdated.csv')
```

```python
df['DateTime'] = pd.to_datetime(df.Date)
df['year'] = df['DateTime'].dt.year
df['month'] = df['DateTime'].dt.month
df['Open'] = df.Open.astype(str)
df['Open'] = df['Open'].str.replace(',', '').astype(float)
df['day'] = df['DateTime'].dt.day
df['30d_avg'] = df['Open'].rolling(window=30, center=False).mean()
df['Std'] = df['Open'].rolling(window=30, center=False).std()
df.head()
```

```python
df.index = df['DateTime']
df.head()
```

```python
opn = df['Open']
plt.figure(figsize = (20,10))
plt.plot(opn)
plt.title("Time Series")
plt.xlabel("Time (year-month)")
plt.ylabel("Open Price")
plt.legend(loc = 'best')
plt.show
```

```python
plt.scatter(df['year'], opn)
plt.show()
```

```python
by_year = df.groupby('year')['Open'].mean()
plt.figure(figsize = (10,5))
by_year.plot.bar(width=0.35)
```

```python
by_month = df.groupby('month')['Open'].mean()
plt.figure(figsize = (12,6))
by_month.plot.bar(width=0.35)
```

```python
by_month = df.groupby('day')['Open'].mean()
plt.figure(figsize = (18,9))
by_month.plot.bar(width=0.35)
```
